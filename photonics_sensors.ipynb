```python
import json
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
%config IPCompleter.greedy=True
mpl.rcParams['agg.path.chunksize'] = 10000
```

```python
def process_plot_results(fname):
    with open(fname, 'r') as stream:
        data = json.load(stream)
    tests_average = []
    for test in data:
        df = pd.DataFrame(test)
        tests_average.append(df.mean())

    df_tests = pd.DataFrame(tests_average)
    
    return df_tests
```

```python
process_plot_results('photonics_sensors_3dist_oldsensor.json').T.plot(marker='o')
plt.savefig('results_oldsensor.png', bbox_inches='tight')
```

```python
process_plot_results('photonics_sensors_3dist_newsensor.json').T.plot(marker='o')
plt.savefig('results_newsensor.png', bbox_inches='tight')
```
